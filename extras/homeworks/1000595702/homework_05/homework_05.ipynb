Task 1

Write your own Simpson method to numerically calculate integral my_simps(some_func,a,b,n)

* some_func
* a: left limit
* b: right limit
* n: number of segments for example:

for any a,b,n, func

```python
import numpy as np
```

```python
def some_func(x):
    return np.exp(x)
```

```python
a=1.5
b=2.5
n=4
```

```python
def my_simps(some_func,a,b,n):
  h = (b - a)/ float(n)
  S = 0.0
  for i in range(1, int(n)):
      x = a + i * h
      if (i%2==0):
          S = S + 2 * some_func(x)
      else:
          S = S + 4 * some_func(x)
  S = S + some_func(a) + some_func(b) 
  result = S * h/3
  return result
```

```python
I=my_simps(some_func, a, b, n)
print(I)
```

Then make a plot for

```python
import matplotlib.pyplot as plt
```

```python
for i in range(0,10):
  print(my_simps(lambda x: i*x**2 + x**2 * np.sqrt(x) - 1,1,9,100))
```

```python
x=np.linspace(1,30,100)
fig,ax=plt.subplots(1,1)

for i in range(0,10):
  ax.plot(x,i*x**2 + x**2 * np.sqrt(x) - 1,label='b = '+str(i))
ax.set_title('Integraci√≥n')
ax.set_xlabel('x') 
ax.set_ylabel('fun(x)') 
ax.legend()
ax.grid()
```
